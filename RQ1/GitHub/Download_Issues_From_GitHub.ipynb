```python
import requests
import json

# Replace 'your_token' with your GitHub token
token = ' '
headers = {'Authorization': 'token {}'.format(token)}

# Specify the owner and repository
owner = ' '
repo = ' '

# Initialize variables
url = 'https://api.github.com/repos/{}/{}'.format(owner, repo) + '/issues'
params = {'state': 'all', 'per_page': 100, 'page': 1}
all_issues = []

# Fetch the issues with pagination
while True:
    response = requests.get(url, headers=headers, params=params)
    issues = response.json()
    for issue in issues:
        # Check if the issue is not a pull request
        if 'pull_request' not in issue:
            all_issues.append(issue)
    if len(issues) < params['per_page']:
        break
    params['page'] += 1

# Write the issues to a JSON file
filename = '{}_{}.json'.format(owner, repo)
with open(filename, 'w') as file:
    json.dump(all_issues, file, indent=4)

print(f'Data has been saved to {filename}')
```

# Knime

```python
import requests
import json

# Replace 'your_token' with your GitHub token
token = ''
headers = {'Authorization': f'token {token}'}

# Specify the owner
owner = ''
base_url = ''

# Fetch all repositories owned by 'galaxyproject'
repos_url = f'{base_url}/orgs/{owner}/repos'
repos_params = {'per_page': 100, 'page': 1}
repos = []

while True:
    response = requests.get(repos_url, headers=headers, params=repos_params)
    repo_data = response.json()
    if not isinstance(repo_data, list):
        print("Unexpected response format:", repo_data)
        break
    repos.extend(repo_data)
    if len(repo_data) < repos_params['per_page']:
        break
    repos_params['page'] += 1

# Fetch issues from each repository
all_issues = []

for repo in repos:
    repo_name = repo['name']
    issues_url = f'{base_url}/repos/{owner}/{repo_name}/issues'
    params = {'state': 'all', 'per_page': 100, 'page': 1}

    while True:
        response = requests.get(issues_url, headers=headers, params=params)
        issues = response.json()
        if not isinstance(issues, list):
            print(f"Unexpected response format for {repo_name}:", issues)
            break
        for issue in issues:
            if 'pull_request' not in issue:
                all_issues.append(issue)
        if len(issues) < params['per_page']:
            break
        params['page'] += 1

# Write the issues to a JSON file
filename = f'{owner}_issues.json'
with open(filename, 'w') as file:
    json.dump(all_issues, file, indent=4)

print(f'Data has been saved to {filename}')
```
