````python
df_combined_issues_prs['title'] = df_combined_issues_prs['title'].apply(
    lambda x: re.sub(r'[^A-Za-z0-9\s.,!?\'"-]', '', str(x))
)
````

````python
df_combined_issues_prs['body'] = df_combined_issues_prs['body'].apply(
    lambda x: re.sub(r'[^A-Za-z0-9\s.,!?\'"-]', '', str(x))
)
````

````python
## Remove code snippets
def remove_code_snippets(text):
    if not text:
        return text
    # Remove inline code (e.g., `code`)
    text = re.sub(r'`[^`]*`', '', text)
    # Remove block code (e.g., ```code```)
    text = re.sub(r'```[\s\S]*?```', '', text)
    return text
````

````python
## Removal of Code Snippets
df_combined_issues_prs['title'] = df_combined_issues_prs['title'].apply(remove_code_snippets)
df_combined_issues_prs['body'] = df_combined_issues_prs['body'].apply(remove_code_snippets)
````

````python
# Function to remove HTTPS links from text
def remove_https_links(text):
    if isinstance(text, str):
        # Regular expression pattern to match HTTPS links
        https_pattern = r'https://\S+'
        # Remove all matches from the text
        return re.sub(https_pattern, '', text).strip()
    else:
        # If not a string, return as is
        return text
````

````python
## Removal of HTTPS Links
df_combined_issues_prs['title'] = df_combined_issues_prs['title'].apply(remove_https_links)
df_combined_issues_prs['body'] = df_combined_issues_prs['body'].apply(remove_https_links)
````

````python
## Function to remove HTML tags
def remove_html_tags(text):
    if text is None:
        return ''
    return re.sub(r'<.*?>', '', text)
````

````python
## Removal of HTML Tags
## Removal of HTTPS Links
df_combined_issues_prs['title'] = df_combined_issues_prs['title'].apply(remove_html_tags)
df_combined_issues_prs['body'] = df_combined_issues_prs['body'].apply(remove_html_tags)
````

````python
df_combined_issues_prs['text'] = df_combined_issues_prs['title'] + ' ' +df_combined_issues_prs['body']
df_combined_issues_prs['text'] 
````

````python
## Removing Stop Words
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
#additional_stopwords=['nextflow', 'perf', 'pr', 'issue', 'pull', 'request', 'precommit', 'nf-core'] ## Set accordingly
#stop_words.update(additional_stopwords)

## Function to remove stopwords
## Remove stop words
def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)
````

````python
## Removing Stop words
df_combined_issues_prs['text'] = df_combined_issues_prs['text'].apply(remove_stopwords)
df_combined_issues_prs
````

````python
## Function to remove special characters
def remove_special_characters(text):
    if text is None:
        return ''
    return re.sub(r'[^a-zA-Z0-9\s]', '', text)
````

````python
## Removing special Character
df_combined_issues_prs['text'] = df_combined_issues_prs['text'].apply(remove_special_characters)
df_combined_issues_prs
````

````python
## Function for Lemmatization
import spacy
nlp = spacy.load('en_core_web_sm')
nltk.download('punkt')
nlp = spacy.load('en_core_web_sm')

def lemmatize_text(text):
    doc = nlp(text)
    lemmatized_tokens = [token.lemma_ for token in doc]
    return ' '. join(lemmatized_tokens)
````

````python
df_combined_issues_prs['text'] = df_combined_issues_prs['text'].apply(lemmatize_text)
#df_nf_core_modules
````
