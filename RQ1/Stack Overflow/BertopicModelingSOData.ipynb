```python
import pandas as pd
```

```python
preprocessed_df = pd.read_csv('Cleaned Data SO.csv')
cols = ['Id', 'Title', 'Text', 'OriginalText']
preprocessed_df = preprocessed_df[cols]
preprocessed_df
```

# Pre-calculate Embeddings

```python
from sentence_transformers import SentenceTransformer

## Precalculate Embedding
embedding_model = SentenceTransformer("BAAI/bge-large-en-v1.5")
```

```python
## Lemmatize Original Text
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

import pandas as pd
import re
from bs4 import BeautifulSoup

from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import spacy

import nltk
nltk.download('punkt')
import spacy
nlp = spacy.load("en_core_web_sm")

def lemmatize_text(text):
    doc = nlp(text if isinstance(text, str) else "")
    return " ".join([t.lemma_ for t in doc])
```

```python
preprocessed_df['OriginalText'] = preprocessed_df['OriginalText'].apply(lemmatize_text)
preprocessed_df['OriginalText']
```

# Dimensionality Reduction

```python
from umap import UMAP
umap_model = UMAP(n_neighbors=10, n_components=5, min_dist=0.1, metric="cosine", random_state=42)  # cosine works well for embeddings
umap_model
```

```python
from hdbscan import HDBSCAN
n_docs = len(preprocessed_df)
min_cluster = max(15, int(0.02 * n_docs))  # ~2% of corpus, floor at 15
hdbscan_model = HDBSCAN(min_cluster_size=min_cluster, metric="euclidean",
                        cluster_selection_method="leaf", prediction_data=True)

hdbscan_model
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer_model = CountVectorizer(stop_words="english", min_df=1, max_df=1.0, ngram_range=(1, 2))
```

```python
## Embedding Generation
embeddings_OriginalText = embedding_model.encode(
    preprocessed_df["OriginalText"],
    show_progress_bar=True,
    normalize_embeddings=True
)
```

```python
# --- BERTopic ---
from bertopic import BERTopic
topic_model = BERTopic(
    embedding_model=embedding_model,
    umap_model=umap_model,
    hdbscan_model=hdbscan_model,
    vectorizer_model=vectorizer_model,
    top_n_words=20,
    verbose=True
)
```

```python
## Topics Generation
# Use the lemmatized text with the vectorizer/BERTopic, but the embeddings from original text are fine
topics, probs = topic_model.fit_transform(preprocessed_df["OriginalText"], embeddings_OriginalText)
topic_model.get_topic_info()
```

```python
topic_model.get_topic_info().to_csv('Generated Topics SO.csv')
```

# Document Distribution

```python
document_distribution = topic_model.get_document_info(preprocessed_df['OriginalText'])
document_distribution
```

```python
concatenated_df = pd.concat([preprocessed_df, document_distribution], axis=1)
concatenated_df
```

```python
distinct_topics = concatenated_df['Topic'].unique()
distinct_topics
```

```python
for topic in distinct_topics:
    df_topic = concatenated_df[concatenated_df['Topic'] == topic]
    filename = f"{topic}_Topics_data.csv"
    df_topic.to_csv(filename)
```
